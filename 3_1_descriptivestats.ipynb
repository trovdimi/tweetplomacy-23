## Tweets distribution

```python
def plot_users(df, title):
    # Plot histogram using pandas
    ax = df.hist(
        column='tweetId',
        bins=15,
        grid=False,
     figsize=(3.33, 2.5),
        zorder=2,
        rwidth=0.9,
        log=True
    )[0][0]  # Access the Axes object from the returned 2D array

    # Set axis labels
    ax.set_xlabel("Number of tweets", fontsize=10)
    ax.set_ylabel("Number of users", fontsize=10)
    
    # Show minor ticks
    ax.minorticks_on()
    ax.yaxis.set_minor_locator(LogLocator(base=10.0, subs='auto', numticks=100))

    # Remove the default title (which is the column name)
    ax.set_title("")
    plt.xticks(fontsize=10, ha='center')
    plt.yticks(fontsize=10, ha='right')
    
    plt.savefig('./plots/userdistribution_'+title+'.pdf', bbox_inches = 'tight')
    plt.tight_layout()
    plt.show()

```

```python
def hashtag_usage_per_topic(df, topic):
    df = df[df[topic]]
 
    return [len(df[df['hashtags'].str.len() != 0]['hashtags'])/len(df),
            len(df[df['hashtags'].str.len() != 0]['hashedUserName'].unique())/len(df.hashedUserName.unique()),
            df[df['hashtags'].str.len() != 0]['hashtags'].apply(len).mean(),    
            df[df['hashtags'].str.len() != 0].groupby('hashedUserName').apply(lambda x : x['hashtags'].apply(len).mean()).mean()]

    
   
```

```python
def hashtag_usage(df):
    hashtagusage={}

    topic='topic_energy_security'
    hashtagusage[topic]= hashtag_usage_per_topic(df, topic)

    topic='topic_covid_19'
    hashtagusage[topic]= hashtag_usage_per_topic(df, topic)

    topic='topic_climate_change'
    hashtagusage[topic]= hashtag_usage_per_topic(df, topic)

    hashtagusage
    return pd.DataFrame.from_dict(hashtagusage, orient='index').transpose()
```

```python
hashtagusage = hashtag_usage(df)
```

```python
print(hashtagusage.to_latex())
```

## Characterize Mention Usage

```python
def mention_usage_per_topic(df, topic):
    df = df[df[topic]]
    
    return [len(df[df['mentions'].str.len() != 0]['mentions'])/len(df),
            len(df[df['mentions'].str.len() != 0]['hashedUserName'].unique())/len(df.hashedUserName.unique()),
            df[df['mentions'].str.len() != 0]['mentions'].apply(len).mean(),
            df[df['mentions'].str.len() != 0].groupby('hashedUserName').apply(lambda x : x['mentions'].apply(len).mean() ).mean()
           ]
    
```

```python
def mention_usage(df):
    mentionusage={}
    topic='topic_energy_security'
    mentionusage[topic]= mention_usage_per_topic(df, topic)

    topic='topic_covid_19'
    mentionusage[topic]= mention_usage_per_topic(df, topic)

    topic='topic_climate_change'
    mentionusage[topic]= mention_usage_per_topic(df, topic)
    
    return pd.DataFrame.from_dict(mentionusage, orient='index').transpose()
```

```python
mentionusage = mention_usage(df)
```

```python
print(mentionusage.to_latex())
```
